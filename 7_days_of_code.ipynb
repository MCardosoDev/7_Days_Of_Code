# Dia 2 Importação de dados

```python
intervals = [0, 99, 199, 299, 399, 499, 599, 699, 799, 899, 999]
categories = ['Generalidades',
              'Filosofia e psicologia',
              'Religião',
              'Ciências sociais',
              'Classe vaga',
              'Matemática e ciências naturais',
              'Ciências aplicadas',
              'Belas artes',
              'Linguagem',
              'Geografia. Biografia. História.']

dataset["CDU_geral"] = pd.cut(dataset['localizacao'], bins=intervals, labels=categories, right=False)
dataset.head(5)
```
